```python
df = pd.DataFrame({'Model': ['Doc-Prism','Doc-Bertscore', 'Doc-COMET', 
                            'Doc-Prism','Doc-Bertscore', 'Doc-COMET'],
                   'Correlation': [0.5949, 0.624385, 0.63007,
                                 0.648585, 0.649585, 0.66347],
                   'Context': ['hypothesis context', 'hypothesis context', 'hypothesis context',
                            'reference context', 'reference context','reference context',]})
```

```python
import seaborn as sns

#set seaborn plotting aesthetics
sns.set(style='whitegrid')


#create grouped bar chart
g = sns.catplot(
    data=df, kind="bar",
    x="Model", y="Correlation", hue="Context",
    ci="sd", palette="dark", alpha=.7)
g.despine(left=True)
# bar = sns.barplot(x='Model', y='Correlation', hue='Context', data=df)
# bar.legend(bbox_to_anchor= (0.16, 1) );
g.set_axis_labels("", "Correlation")
g._legend.remove()
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.07),
          fancybox=True, shadow=True, ncol=5)
```

```python
g.savefig('context_quality.png', dpi=300, bbox_inches='tight')
```
